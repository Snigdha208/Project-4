```python
# Convert 'AREA (1000 ha)' to 'AREA (ha)'
for col in df.columns:
    if 'AREA (1000 ha)' in col:
        df[col] = df[col] * 1000
        df.rename(columns={col: col.replace(' (1000 ha)', ' (ha)')}, inplace=True)

# Convert 'PRODUCTION (1000 tons)' to 'PRODUCTION (tons)'
for col in df.columns:
    if 'PRODUCTION (1000 tons)' in col:
        df[col] = df[col] * 1000
        df.rename(columns={col: col.replace(' (1000 tons)', ' (tons)')}, inplace=True)

for col in df.columns:
    if 'YIELD (Kg per ha)' in col:
        df[col] = df[col] / 1000  # kg to tons
        df.rename(columns={col: col.replace(' (Kg per ha)', ' (tons per ha)')}, inplace=True)       
```

```python
# Make all column names lowercase and replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
import re
def clean_column(col):
    col = col.strip().lower()
    col = re.sub(r'\(tons_per_ha\)', 'tph', col)
    col = re.sub(r'\(tons\)', 'tons', col)
    col = re.sub(r'\(ha\)', 'ha', col)
    col = re.sub(r'[()\s]', '', col)  # remove leftover () or extra spaces
    col = col.replace("__", "_")      # replace double underscores
    return col

# Apply to all columns
df.columns = [clean_column(col) for col in df.columns]

# To verify:
print(df.columns)
```

```python
negatives = {}
for col in df.select_dtypes(include=['number']).columns:
    neg_count = (df[col] < 0).sum()
    if neg_count > 0:
        negatives[col] = neg_count

print("Columns with negative values and their counts:")
print(negatives)
```

```python
import numpy as np

# Replace all negative values in numeric columns with NaN
df[df.select_dtypes(include=['number']).columns] = df.select_dtypes(include=['number']).applymap(
    lambda x: np.nan if x < 0 else x
)

# Check columns with missing values after replacement
missing_counts = df.isna().sum()
print("Missing values per column (after replacement):")
print(missing_counts[missing_counts > 0])
```

```python
#Median is robust to outliers (unlike mean), which is ideal in agricultural data where yields can vary widely.

# Impute missing values in numeric columns with median
numeric_cols = df.select_dtypes(include=['number']).columns

# Replace NaN with median for each column
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Confirm no missing values remain
missing_after_impute = df.isna().sum()
print("Missing values per column (after imputation):")
print(missing_after_impute[missing_after_impute > 0])
```

```python
# SQL QUERIES

# 1. Year-wise Trend of Rice Production Across States (Top 3)

query1 = """
    -- Step 1: Get top 3 states by total rice production
    WITH top_states AS (
        SELECT state_name
        FROM crop_production
        GROUP BY state_name
        ORDER BY SUM(rice_production_tons) DESC
        LIMIT 3
    )

    -- Step 2: Year-wise rice production for those top states
    SELECT year, state_name, SUM(rice_production_tons) AS total_production
    FROM crop_production
    WHERE state_name IN (SELECT state_name FROM top_states)
    GROUP BY year, state_name
    ORDER BY year, state_name;
"""
# Run the query and load into a DataFrame
df_query1 = pd.read_sql(query1, conn)

# Print the DataFrame
df_query1.head(5)
```

```python
# 2.Top 5 Districts by Wheat Yield Increase Over the Last 5 Years

query2 = '''
WITH WheatYield AS (
    SELECT
        dist_name,
        year,
        AVG(wheat_yield_tpha) AS avg_yield
    FROM crop_production
    WHERE year IN (2012, 2017)
    GROUP BY dist_name, year
),
YieldComparison AS (
    SELECT
        curr.dist_name,
        curr.avg_yield AS yield_recent,
        prev.avg_yield AS yield_5years_ago,
        (curr.avg_yield - prev.avg_yield) AS yield_increase
    FROM WheatYield curr
    JOIN WheatYield prev ON curr.dist_name = prev.dist_name
    WHERE curr.year = 2017 AND prev.year = 2012
)
SELECT *
FROM YieldComparison
ORDER BY yield_increase DESC
LIMIT 5;
'''

df_query2 = pd.read_sql(query2, conn)
df_query2
```

```python
# 3. States with the Highest Growth in Oilseed Production (5-Year Growth Rate)
query3 = """
WITH oilseed_data AS (
    SELECT state_name AS state, year, SUM(oilseeds_production_tons) AS total_production
    FROM crop_production
    GROUP BY state_name, year
),
prod_diff AS (
    SELECT 
        state,
        MAX(CASE WHEN year = 2012 THEN total_production END) AS prod_5yrs_ago,
        MAX(CASE WHEN year = 2017 THEN total_production END) AS prod_now
    FROM oilseed_data
    GROUP BY state
)
SELECT 
    state,
    prod_5yrs_ago,
    prod_now,
    ROUND(((prod_now - prod_5yrs_ago) / prod_5yrs_ago) * 100, 2) AS growth_rate_percent
FROM prod_diff
WHERE prod_5yrs_ago IS NOT NULL AND prod_now IS NOT NULL
ORDER BY growth_rate_percent DESC
LIMIT 10;
"""

df_query3 = pd.read_sql(query3, conn)
df_query3
```

```python
# 4. District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize)

# List of crops and their corresponding area and production columns
crops = {
    'rice': ('rice_area_ha', 'rice_production_tons'),
    'wheat': ('wheat_area_ha', 'wheat_production_tons'),
    'maize': ('maize_area_ha', 'maize_production_tons')
}

# Results list to store correlation per district for each crop
results = []

# Iterate through each crop
for crop, (area_col, prod_col) in crops.items():
    # Filter relevant columns and drop NA
    crop_df = df[['dist_name', area_col, prod_col]].dropna()
    
    # Filter out non-positive values (to avoid invalid correlations)
    crop_df = crop_df[(crop_df[area_col] > 0) & (crop_df[prod_col] > 0)]
    
    # Group by district
    for dist, group in crop_df.groupby('dist_name'):
        if len(group) >= 2:  # At least 2 data points required for correlation
            corr = group[area_col].corr(group[prod_col])
            results.append({
                'district': dist,
                'crop': crop,
                'correlation': round(corr, 3)
            })

# Convert results to DataFrame
df_query4 = pd.DataFrame(results)

# Display results
print("District-wise Correlation Between Area and Production for Major Crops:")
df_query4.head(10)  # show top 10 rows for quick look
```

```python
# 5. Yearly Production Growth of Cotton in Top 5 Cotton Producing States

query5 = """
-- Step 1: Get top 5 cotton producing states overall
WITH top_states AS (
    SELECT state_name
    FROM crop_production
    GROUP BY state_name
    ORDER BY SUM(cotton_production_tons) DESC
    LIMIT 5
)

-- Step 2: Get year-wise cotton production for those states
SELECT 
    year, 
    state_name, 
    SUM(cotton_production_tons) AS total_cotton_production
FROM crop_production
WHERE state_name IN (SELECT state_name FROM top_states)
GROUP BY year, state_name
ORDER BY year, state_name;
"""

# Run the query
df_query5 = pd.read_sql(query5, conn)

# Display result
print("Year-wise Cotton Production in Top 5 Producing States:")
df_query5.head(10)
```

```python
# 7. Annual Average Maize Yield Across All States

query7 = """
SELECT 
    year,
    ROUND(AVG(maize_yield_tpha), 2) AS avg_maize_yield_tpha
FROM crop_production
GROUP BY year
ORDER BY year;
"""

df_query7 = pd.read_sql(query7, conn)
print("Annual Average Maize Yield Across All States:")
df_query7.head(5)
```

```python
# 8. Total Area Cultivated for Oilseeds in Each State
query8 = """
SELECT 
    state_name, 
    ROUND(SUM(oilseeds_area_ha), 2) AS total_oilseed_area_ha
FROM crop_production
GROUP BY state_name
ORDER BY total_oilseed_area_ha DESC;
"""

df_query8 = pd.read_sql(query8, conn)
print("Total Oilseed Area by State:")
df_query8.head(5)
```

```python
# 9. Districts with the Highest Rice Yield
query9 = """
SELECT 
    dist_name, 
    state_name, 
    ROUND(AVG(rice_yield_tpha), 2) AS avg_rice_yield
FROM crop_production
GROUP BY dist_name, state_name
ORDER BY avg_rice_yield DESC
LIMIT 10;
"""

df_query9 = pd.read_sql(query9, conn)
print("Districts with the Highest Average Rice Yield:")
df_query9
```

```python
# 10. Compare the Production of Wheat and Rice for the Top 5 States Over 10 Years

query10 = """
-- Get top 5 states by combined wheat + rice production
WITH top_states AS (
    SELECT state_name
    FROM crop_production
    GROUP BY state_name
    ORDER BY SUM(wheat_production_tons + rice_production_tons) DESC
    LIMIT 5
)

-- Yearly wheat and rice production in those states
SELECT 
    year, 
    state_name,
    SUM(wheat_production_tons) AS total_wheat,
    SUM(rice_production_tons) AS total_rice
FROM crop_production
WHERE state_name IN (SELECT state_name FROM top_states)
GROUP BY year, state_name
ORDER BY year, state_name;
"""

df_query10 = pd.read_sql(query10, conn)
print("Wheat vs Rice Production in Top 5 States Over 10 Years:")
df_query10.head(10)
```
